```python
# Load raw tourism data
raw_data = pd.read_csv('raw_data.csv', sep='\t', decimal=',', encoding="ISO-8859-1")
print("Initial data shape:", raw_data.shape)
raw_data.head()
```

```python
# Remove aggregated total row
tourism_data = raw_data[raw_data['País de residencia'] != 'Total']
print("Shape after removing Total row:", tourism_data.shape)
tourism_data.head()
```

```python
# Check for data quality
tourism_data.info()
```

```python
# Clean data: handle missing values and format numbers
tourism_data = tourism_data.dropna(subset=['Total'])
tourism_data['Total'] = pd.to_numeric(tourism_data['Total'].str.replace('.', ''), errors='coerce')
print("Data shape after cleaning:", tourism_data.shape)
tourism_data.head()
```

```python
# Create time series format (for area chart)
timeseries_data = tourism_data.pivot(
    index='Periodo',
    columns='País de residencia',
    values='Total'
)
timeseries_data = timeseries_data.reset_index()
print("Time series data shape:", timeseries_data.shape)
timeseries_data.head()
```

```python
# Save time series data
timeseries_data.to_csv('area_chart.csv', index=False)
```

```python
# Extract month and drop unused columns
tourism_data['Month'] = tourism_data['Periodo'].str[-2:]
tourism_data = tourism_data.drop(['Municipio de destino', "Concepto turístico", "Periodo"], axis=1)
tourism_data.head()
```

```python
# Calculate average monthly visitors per country
monthly_visitors = (tourism_data.groupby(['País de residencia', 'Month'])['Total']
                   .mean()
                   .round()
                   .unstack())
```

```python
# Set proper month labels
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_visitors.columns = month_names
print("Monthly visitors shape:", monthly_visitors.shape)
monthly_visitors.head()
```

```python
# Calculate total visitors to Barcelona by country of origin
visitors_by_country = monthly_visitors.sum(axis=1).sort_values(ascending=False)
top_source_countries = visitors_by_country.head(3).index.tolist()
print("Top 3 countries of origin for Barcelona tourists:")
print(visitors_by_country.head(3))
```

```python
# Prepare regional analysis data (for beeswarm plot)
regional_data = pd.DataFrame({
    'turistas': visitors_by_country,
    'región': pd.Series(country_to_region)
}).reset_index()

regional_data.columns = ['país', 'turistas', 'región']
regional_data = regional_data[['región', 'país', 'turistas']].astype({'turistas': int})
print("Tourist origins by region:")
regional_data.head(10)
```

```python
# Save regional data
regional_data.to_csv('beeswarm.csv', index=False)
```

```python
# Prepare monthly comparison of visitors from top source countries (for Nightingale plot)
monthly_comparison = monthly_visitors.loc[top_source_countries].T.reset_index()
monthly_comparison.columns = ['month', 'france', 'uk', 'usa']
monthly_comparison = monthly_comparison.astype({'france': int, 'uk': int, 'usa': int})
print("Monthly visitors from top source countries:")
monthly_comparison
```

```python
# Save monthly comparison data
monthly_comparison.to_csv('nightingale.csv', index=False)
```
